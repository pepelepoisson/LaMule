```python
# Sous-routine servant à préparer et formater rapidement des séries de graphs x*y servant à l'exploration des données

from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
# Define the date format
#date_form = DateFormatter("%m-20%y")

# Subroutine used to create a grid of XY plots
def grid_scatterplot(pgda_data, x_parameter, y_parameters, color_parameter, title):
    import matplotlib.pyplot as plt

    # Definition of the grid size used to display results based on the number of parameters selected
    grid_x=3
    if (len(y_parameters.axes[1])%grid_x==0):
        grid_y=len(y_parameters.axes[1])//grid_x
    else:
        grid_y=len(y_parameters.axes[1])//grid_x+1
    
    # Definition of the plots - Loops over the parameters contained in y_parameters
    fig, axs = plt.subplots(grid_y, grid_x, figsize=(grid_x*4,grid_y*4))
    fig.tight_layout(h_pad=3)
    fig.suptitle(title)
    plt.subplots_adjust(hspace =0.2, top=0.95, wspace=0.5)

    for i in range(0, len(y_parameters.axes[1])):
        colname = y_parameters.columns[i]
        x_pos=(i)%grid_x
        y_pos=(i)//grid_x
        ax = sns.scatterplot(x=x_parameter, y=pgda_data[colname], hue=color_parameter, ax=axs[y_pos,x_pos], palette="deep")
        #scatter.legend(fontsize = 12,bbox_to_anchor= (1.03, 1),title=color_parameter.name, title_fontsize = 12,shadow = True,facecolor = 'white');
        #ax.set(xlabel="Date", ylabel="Calculated inlet losses (inH2O)", title=site_id)
        #ax.xaxis.set_major_formatter(date_form)
        # Ensure ticks fall once every 52 weeks
        #ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=52))
```

```python
# La resolution de l'enregistrement d'altitude est faible (1cm). Ne garder  qu'un point sur 10
Data=combined_datalog.iloc[::1, :].copy()
#display(Data.head())

# Calcul de la vitesse verticale par dérivation de l'altitude lissée
Data_diff=Data[['time_sec','altitude','altitude_lissee','alt_lissee','alt']].diff().copy()
Data['delta_sec']=Data_diff.time_sec  # Pas de temps
Data['vitesse_verticale']=Data_diff.alt_lissee/Data.delta_sec
Data['vitesse_verticale_km/h']=Data['vitesse_verticale']*3.6

#Data['detected_phase']="APV"

#for i in ['vol 5', 'vol 6', 'vol 7']:
for i in ['vol 4','vol 5', 'vol 6', 'vol 7']:
    print(">>>>>>>> "+i)
    df=Data[Data.source==i].copy()
    df.reset_index(drop=True, inplace=True)
    size=df.shape[0]
    detected_phase=["0: Attente Pre Vol"]*size
    
    # Détéction MECO
    time_motor_burnout=df[df.vitesse_verticale == df.vitesse_verticale.max()]['time_sec'].values[0]
    alt_motor_burnout=df[df.vitesse_verticale == df.vitesse_verticale.max()]['altitude'].values[0]
    speed_motor_burnout=df[df.vitesse_verticale == df.vitesse_verticale.max()]['vitesse_verticale'].values[0]
    print("Motor burnout: "+str(round(time_motor_burnout,2))+"s "+str(round(alt_motor_burnout,2))+"m "+str(round(speed_motor_burnout,2))+"m/s")
    
    # Détéction Apogée
    alt_apogee=df[['time_sec', 'alt_lissee','vitesse_verticale_km/h']][df.source==i].alt_lissee.max()
    time_apogee=df[(df.alt_lissee==alt_apogee)]['time_sec'].min()
    print("Apogée: "+str(round(time_apogee,2))+"s "+str(round(alt_apogee,2))+"m")

    # Phase 0: Attente_pre-vol (APV)
    # Evenement 1: Décollage
    # Phase 1: vol_propulsé (VP)
    # Evenement 2: MECO
    # Phase 2: vol_ballistique_ascendant (VBA)
    # Evenement 3: Apogée
    # Phase 3: vol_ballistique_descendant (VBD)
    # Evenement 4: Atterrissage
    # Phase 4: attente_après-vol (AAV)
    lag=100
    take_off_altitude_detect=0.1
    take_off_altitude_check=0
    take_off_accel_detect=5
    take_off_accel_check=0
    if (df['vertical_acceleration'].isnull().values.any()):
        take_off_accel_check=1
    post_apogee_altitude_detect=0.25
    landed_altitude_detect=0.05
    phase="0: Attente Pre Vol"
    lagged_ind=0
    for ind in df.index:
        if (df['millis'][ind]-df['millis'][lagged_ind]>lag):
            if((df.altitude[ind]-df.altitude[lagged_ind]>=take_off_altitude_detect)&(take_off_altitude_check==0)):
                take_off_altitude_check=1
                print("alt shift time: "+str(df.time_sec[ind]))
            if((df.vertical_acceleration[ind]-df.vertical_acceleration[lagged_ind]>=take_off_accel_detect)&(take_off_accel_check==0)):
                take_off_accel_check=1
                print("accel shift time: "+str(df.time_sec[ind]))
            if((take_off_altitude_check*take_off_accel_check==1)&(phase=="0: Attente Pre Vol")):
                take_off_time=df.time_sec[ind]
                phase="1: Vol Propulsé"
                print("take off time: "+str(take_off_time))
            if((phase=="1: Vol Propulsé")&((df.vertical_acceleration[ind]<=0)|(df.vitesse_verticale[ind] == df.vitesse_verticale.max()))):
                meco_time=df.time_sec[ind]
                phase="2: Vol Ballistique Ascendant"
                print("MECO time: "+str(meco_time))
            if((phase=="2: Vol Ballistique Ascendant")&(df.max_altitude[ind]-df.altitude_lissee[ind]>=post_apogee_altitude_detect)):
                apogee_time=df.time_sec[ind]
                phase="3: Vol Ballistique Descendant"
                print("apogee time: "+str(apogee_time))
            if((phase=="3: Vol Ballistique Descendant")&(df.altitude_lissee[lagged_ind]-df.altitude_lissee[ind]<=landed_altitude_detect)&(df.max_altitude[ind]-df.altitude_lissee[ind]>10*post_apogee_altitude_detect)):
                landing_time=df.time_sec[ind]
                phase="4: Attente Post Atterrissage"
                print("landing time: "+str(landing_time))
            detected_phase[ind]=phase
            lagged_ind=lagged_ind+1

    df['detected_phase']=detected_phase
    if (i=='vol 4'):
        updated_combined_datalog=df.copy()
    if (i!='vol 4'):
        updated_combined_datalog = pd.concat([updated_combined_datalog, df], ignore_index=True)

    print("")

dict_palette = {
    '0: Attente Pre Vol': 'tab:blue',
    '1: Vol Propulsé': 'tab:green',
    '2: Vol Ballistique Ascendant': 'tab:orange',
    '3: Vol Ballistique Descendant': 'tab:red',
    '4: Attente Post Atterrissage': 'tab:purple'}
PlottingData=updated_combined_datalog.copy()
plot=sns.scatterplot(x=PlottingData['time_sec'], y=PlottingData['alt_lissee'], hue=PlottingData.detected_phase, s=6, palette=dict_palette, linewidth=0)
plot.set(title='Altitude et phase de vol - LaMule vols 4, 5, 6 et 7 avec moteurs D12-5')
plot.set_xlabel("Temps écoulé depuis le décollage (secondes)", fontsize = 12)
plot.set_ylabel("Altitude (m)", fontsize = 12)
```

```python
plot=sns.scatterplot(x=PlottingData['time_sec'], y=PlottingData['vitesse_verticale_km/h'], hue=PlottingData.detected_phase, s=6, palette=dict_palette, linewidth=0)
plot.set(title='Vitesses verticales selon phases de vol - LaMule vols 4, 5, 6 et 7 avec moteurs D12-5')
plot.set_xlabel("Temps écoulé depuis le décollage (secondes)", fontsize = 12)
plot.set_ylabel("Vitesse verticale (km/h)", fontsize = 12)
```
