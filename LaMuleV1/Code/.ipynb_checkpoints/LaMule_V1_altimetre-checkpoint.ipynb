## Step 0: Subroutines definition
This section (Step 0) is intended to define commonly used subroutines. Recommend just checking the top comments (green text) to find-out what subroutnes are available. No need to read code in details unless you plan to make some changes.

```python
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
# Define the date format
#date_form = DateFormatter("%m-20%y")

# Subroutine used to create a grid of XY plots
def grid_scatterplot(pgda_data, x_parameter, y_parameters, color_parameter, title):
    import matplotlib.pyplot as plt

    # Definition of the grid size used to display results based on the number of parameters selected
    grid_x=3
    if (len(y_parameters.axes[1])%grid_x==0):
        grid_y=len(y_parameters.axes[1])//grid_x
    else:
        grid_y=len(y_parameters.axes[1])//grid_x+1
    
    # Definition of the plots - Loops over the parameters contained in y_parameters
    fig, axs = plt.subplots(grid_y, grid_x, figsize=(grid_x*4,grid_y*4))
    fig.tight_layout(h_pad=3)
    fig.suptitle(title)
    plt.subplots_adjust(hspace =0.2, top=0.95, wspace=0.5)

    for i in range(0, len(y_parameters.axes[1])):
        colname = y_parameters.columns[i]
        x_pos=(i)%grid_x
        y_pos=(i)//grid_x
        ax = sns.scatterplot(x=x_parameter, y=pgda_data[colname], hue=color_parameter, ax=axs[y_pos,x_pos], palette="deep")
        #scatter.legend(fontsize = 12,bbox_to_anchor= (1.03, 1),title=color_parameter.name, title_fontsize = 12,shadow = True,facecolor = 'white');
        #ax.set(xlabel="Date", ylabel="Calculated inlet losses (inH2O)", title=site_id)
        #ax.xaxis.set_major_formatter(date_form)
        # Ensure ticks fall once every 52 weeks
        #ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=52))
```

## Post vol 3 LaMule
Malheureusement le module d'enregistrement sur carte SD était mal connecté et n'a donc rien enregistré... Frustrant! L'atitude maximale atteinte a peut-etre tout de meme été enregristrée dans la mémoire EEPROM du microcontroleur. Vérifions avec un enregistrement post vol 3...

```python
# Create a dataframe to store solver inputs and outputs from the corresponding *xls file then merge into a single one. 
# Repeat for all units.
Data = pd.read_csv('LOG00006.TXT')

# Add proper date field
#Data['date'] = pd.to_datetime(Data['value_timestamp'])

# Prevew results
print("Data: "+str(Data.shape))
Data.tail()

Data['EEPROM_max']=Data['max_altitude_dm']/10

print('Data aquisition frequency: '+str(round(Data.shape[0]/((Data.millis.max()-Data.millis.min())/1000),2))+' Hz')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

plot=sns.lineplot(x=Data['millis'], y=Data['altitude_lissee'], label='Altitude lissée',color='red', linewidth=2.5)
sns.lineplot(x=Data['millis'], y=Data['max_altitude'], label='Altitude maximale',color='green', linewidth=2.5)
sns.lineplot(x=Data['millis'], y=Data['descente_detectee'], label='Descente détéctée',color='orange', linewidth=2.5)
#sns.lineplot(x=Data['millis'], y=Data['EEPROM_max'], label='Altitude EEPROM',color='pink', linewidth=2.5)
sns.scatterplot(x=Data['millis'], y=Data['altitude'], label='Altitude instantanée')
plot.set_xlabel("Time steps", fontsize = 12)
plot.set_ylabel("Altitude (m)", fontsize = 12)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
PlotData=Data
plot=sns.scatterplot(x=PlotData['millis'], y=PlotData['EEPROM_max'], label='Altitude maximale memoire EEPROM').set(title='Lecture altitude maximale dans la mémoire EEPROM')
```

Yes!!!!!!!!! Cela a bien fonctionné! L'altitude maximale du vol 3 était de 170.2m! La prévision OpenRocket était 195m. La fusée a atteri à 117 pas (environ 117m) du point de décollage. Langle de tir était d'environ 5 degrés.

## Post vol 4 LaMule

```python
# Create a dataframe to store solver inputs and outputs from the corresponding *xls file then merge into a single one. 
# Repeat for all units.
Data = pd.read_csv('LOG00009.TXT')

# Add proper date field
#Data['date'] = pd.to_datetime(Data['value_timestamp'])

# Prevew results
print("Data: "+str(Data.shape))
Data.tail()

Data['EEPROM_max']=Data['max_altitude_dm']/10

print('Data aquisition frequency: '+str(round(Data.shape[0]/((Data.millis.max()-Data.millis.min())/1000),2))+' Hz')
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

plot=sns.lineplot(x=Data['millis'], y=Data['altitude_lissee'], label='Altitude lissée',color='red', linewidth=2.5)
sns.lineplot(x=Data['millis'], y=Data['max_altitude'], label='Altitude maximale',color='green', linewidth=2.5)
sns.lineplot(x=Data['millis'], y=Data['descente_detectee'], label='Descente détéctée',color='orange', linewidth=2.5)
#sns.lineplot(x=Data['millis'], y=Data['EEPROM_max'], label='Altitude EEPROM',color='pink', linewidth=2.5)
sns.scatterplot(x=Data['millis'], y=Data['altitude'], label='Altitude instantanée')
plot.set_xlabel("Time steps", fontsize = 12)
plot.set_ylabel("Altitude (m)", fontsize = 12)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
PlotData=Data[(Data.millis>180100)&(Data.millis<182000)]
plot=sns.lineplot(x=PlotData['millis'], y=PlotData['altitude_lissee'], label='Altitude lissée',color='red', linewidth=2.5)
sns.lineplot(x=PlotData['millis'], y=PlotData['max_altitude'], label='Altitude maximale',color='green', linewidth=2.5)
sns.lineplot(x=PlotData['millis'], y=PlotData['descente_detectee'], label='Descente détéctée',color='orange', linewidth=2.5)
#sns.lineplot(x=Data['millis'], y=Data['EEPROM_max'], label='Altitude EEPROM',color='pink', linewidth=2.5)
sns.scatterplot(x=PlotData['millis'], y=PlotData['altitude'], label='Altitude instantanée')
plot.set_xlabel("Time steps", fontsize = 12)
plot.set_ylabel("Altitude (m)", fontsize = 12)
```

```python
offset_time=180700
offset_alt=Data[(Data.millis>170000)&(Data.millis<180000)]['altitude_lissee'].mean()
Data['time_sec']=(Data['millis']-offset_time)/1000
Data['Altitude']=Data.altitude-offset_alt
Data['Altitude_Lissee']=Data.altitude_lissee-offset_alt
Data['Max_Altitude']=Data.max_altitude-offset_alt
Data['Descente']=Data.descente_detectee*100

# Ajout d'étiquettes à certaines données
temps_decollage=Data[Data.time_sec>=0]['time_sec'].min()
Data.loc[Data['time_sec'] == temps_decollage,'note']='Décollage'
print(Data[Data.time_sec>=0].head())
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
PlotData=Data[(Data.time_sec>-1)&(Data.time_sec<=40)]
plot=sns.lineplot(x=PlotData['time_sec'], y=PlotData['Altitude_Lissee'], label='Altitude lissée',color='red', linewidth=2.5)
sns.lineplot(x=PlotData['time_sec'], y=PlotData['Max_Altitude'], label='Altitude maximale',color='green', linewidth=2.5)
sns.lineplot(x=PlotData['time_sec'], y=PlotData['Descente'], label='Descente détéctée',color='orange', linewidth=2.5)
#sns.lineplot(x=Data['millis'], y=Data['EEPROM_max'], label='Altitude EEPROM',color='pink', linewidth=2.5)
sns.scatterplot(x=PlotData['time_sec'], y=PlotData['Altitude'], label='Altitude instantanée')
plot.set_xlabel("Time steps", fontsize = 12)
plot.set_ylabel("Altitude (m)", fontsize = 12)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
PlotData=Data[(Data.time_sec>6.5)&(Data.time_sec<=7.5)]
plot=sns.lineplot(x=PlotData['time_sec'], y=PlotData['Altitude_Lissee'], label='Altitude lissée',color='red', linewidth=2.5)
sns.lineplot(x=PlotData['time_sec'], y=PlotData['Max_Altitude'], label='Altitude maximale',color='green', linewidth=2.5)
#sns.lineplot(x=PlotData['time_sec'], y=PlotData['Descente'], label='Descente détéctée',color='orange', linewidth=2.5)
#sns.lineplot(x=Data['millis'], y=Data['EEPROM_max'], label='Altitude EEPROM',color='pink', linewidth=2.5)
sns.scatterplot(x=PlotData['time_sec'], y=PlotData['Altitude'], label='Altitude instantanée')
plot.set_xlabel("Time steps", fontsize = 12)
plot.set_ylabel("Altitude (m)", fontsize = 12)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
PlotData=Data
plot=sns.scatterplot(x=PlotData['millis'], y=PlotData['temperature'], label='Temperature ( C )').set(title='Temperature du module barométrique')
```

```python
Data_diff=Data.diff()
Data['delta_sec']=Data_diff.time_sec
Data['vitesse_verticale']=Data_diff.altitude_lissee/Data.delta_sec
PlotData=Data[(Data.time_sec>-1)&(Data.time_sec<=40)]
ax=sns.scatterplot(x=PlotData['time_sec'], y=PlotData['vitesse_verticale'], label='vitesse instantanée')
ax.set(ylim=(-25, 100))
```

```python
lambda_coef=0.1
size=Data.shape[0]
vitesse_verticale_lissee=[0]*size
for i in range(1,size):
    vitesse_verticale_lissee[i]=Data.vitesse_verticale[i]*lambda_coef+Data.vitesse_verticale_lissee[i-1]*(1-lambda_coef)
    
Data['vitesse_verticale_lissee']=vitesse_verticale_lissee

PlotData=Data[(Data.time_sec>-1)&(Data.time_sec<=40)]
ax=sns.scatterplot(x=PlotData['time_sec'], y=PlotData['vitesse_verticale'], label='vitesse verticale instantanée')
ax=sns.scatterplot(x=PlotData['time_sec'], y=PlotData['vitesse_verticale_lissee'], label='vitesse verticale lissee')
ax.set(ylim=(-25, 100))
```

```python
Data_diff2=Data.diff()
Data['acceleration_verticale']=Data_diff2.vitesse_verticale_lissee/Data.delta_sec
PlotData=Data[(Data.time_sec>-1)&(Data.time_sec<=40)]
ax=sns.scatterplot(x=PlotData['time_sec'], y=PlotData['acceleration_verticale'], label='acceleration verticale')
ax.set(ylim=(-100, 200))
```
